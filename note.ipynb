```python
def generate_v(n, seed):
    np.random.seed(seed)
    v = np.random.rand(n).astype('double')
    for i in range(v.shape[0]):
        print(f"v[{i}] = {v[i]};")
    return v
```

```python
def arnoldi_iteration(A, b, m: int):
    """Computes a basis of the (n + 1)-Krylov subspace of A: the space
    spanned by {b, Ab, ..., A^n b}.

    Arguments
      A: m × m array
      b: initial vector (length m)
      n: dimension of Krylov subspace, must be >= 1
    
    Returns
      Q: m x (n + 1) array, the columns are an orthonormal basis of the
        Krylov subspace.
      h: (n + 1) x n array, A on basis Q. It is upper Hessenberg.  
    """
    eps = 1e-12
    h = np.zeros((m+1,m))
    Q = np.zeros((A.shape[0],m+1))
     # Normalize the input vector
    Q[:,0] =b/np.linalg.norm(b,2)   # Use it as the first Krylov vector
    for k in range(1,m+1):
        v = np.dot(A,Q[:,k-1])  # Generate a new candidate vector
        for j in range(k):  # Subtract the projections on previous vectors
            h[j,k-1] = np.dot(Q[:,j].T, v)
            v = v - h[j,k-1] * Q[:,j]
        h[k,k-1] = np.linalg.norm(v,2)
        if h[k,k-1] > eps:  # Add the produced vector to the list, unless
            Q[:,k] = v/h[k,k-1]
        else:  # If that happens, stop iterating.
            return Q, h
    return Q, h
```

```python
def expm_multiply_ground_truth(A, v):
    return scipy.linalg.expm(A) @ v
#     return scipy.sparse.linalg.expm_multiply(A, v)
```

```python
def arnoldi_cheby_expm_multiply(A, v, m):
    Q, H = arnoldi_iteration(A, v, m)
#     print(f"H: {H.shape}")
#     for i in range(H.shape[0]):
#         print(H[i])
#     print(f"\nQ: {Q.shape}")
#     for i in range(Q.shape[0]):
#         print(Q[i])
    Hm = H[:-1,:].astype(complex)
    Qm = Q[:,:-1].astype(complex)
#     print(scipy.linalg.expm(Hm))
#     print(f"Hm: {Hm.shape}")
#     print(Hm)
#     print(f"Qm: {Qm.shape}")
#     print(Qm)
    e1 = np.zeros((m, 1)).astype(complex)
    e1[0] = 1
#     print(e1)
    y = np.zeros((m, 1), dtype=complex)
    y += cheby_coefs[0] * e1
#     print(y)
    Im = np.eye(m, dtype=complex)
#     print(Im)
    for i in range(n_root):
        H_minus_theta_I = -Hm - Im * cheby_roots[i]
        yi = scipy.linalg.solve(H_minus_theta_I, e1)
        y += yi * cheby_coefs[i+1]
    eAv_arnoldi_cheby = (np.linalg.norm(v) * Qm.real @ y.real).flatten()
    eAv_arnoldi = (np.linalg.norm(v) * Qm.real @ scipy.linalg.expm(Hm.real) @ e1.real).flatten()
    return eAv_arnoldi_cheby
```

```python
A, v = generate_A_v(50, 76, 2379)
eAv_true = expm_multiply_ground_truth(A, v)
eAv_arnoldi_cheby = arnoldi_cheby_expm_multiply(A, v, 11)
err = np.abs((eAv_true - eAv_arnoldi_cheby) / eAv_true)
print(err)
```

```python
def test_arnoldi_cheby(n):
    A, v = generate_A_v(n, 76, 2379)
    eAv_true = expm_multiply_ground_truth(A, v)
#     print(f"ground truth:\t{eAv_true}")
    ms = np.array(range(1, n*2+1))
    errors = []
    for m in ms:
        eAv_arnoldi_cheby = arnoldi_cheby_expm_multiply(A, v, m)
        err = np.linalg.norm(np.abs((eAv_true - eAv_arnoldi_cheby) / eAv_true))
#         print(f"m={m}:\t\t{eAv_arnoldi_cheby.flatten()}\terror: {err}")
        errors.append(err)
    plt.plot(ms, np.array(errors))
    print(errors)
```
